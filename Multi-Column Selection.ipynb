```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/home/imaad/products.csv")
batches_df = pd.read_csv("/home/imaad/batches.csv")
```

```python
df['shipping_date'] = pd.to_datetime(df.shipping_date)
```

```python
df
```

```python
batches_df['start_date'] = pd.to_datetime(batches_df.start_date)
batches_df['end_date'] = pd.to_datetime(batches_df.end_date)

batches_df
```

```python
def add_batch_id(x):
    batch_df = batches_df.apply(lambda y: y['batch_id'] if 
                                (x['shipping_date'] >= y['start_date'] and x['shipping_date']<= y['end_date']
                                                and x['product_id'] == y['product_id'])
                                                else None, axis=1)
    batch_df = batch_df.dropna(axis=0, how='all')  
    if batch_df.empty:
        batch_df= np.nan
    else: 
        batch_df= batch_df.to_string(index=False)
        
    return batch_df
    

    
df['batch_id'] = df.apply(add_batch_id, axis=1)
```
